```python
from PIL import Image
from pathlib import Path
from tifffile import TiffFile
import numpy as np
import napari
```

```python
#Specify data path
file = Path(r"path_to_your_data_here") #If you have an acquisition stack divided into multiple files, loop through them so you can load them all together
tif = TiffFile(file)
```

```python
#Load .tiff files and turn into numpy array
img_list = []

for page in tif.pages:
    img_list.append(page.asarray())
    
img_arr = np.array(img_list)
img_arr.shape
```

```python
#Open raw data
viewer = napari.Viewer()
viewer.add_image(img_arr)
```

```python
#Specify cropping (Awful way to reshape array, but works)
#Since we are not triggering acquisition and scanning together, there will be a missmatch between the two at the start
start_frame = 20 #At what frame does the first full scanned volume start
frames_per_vol = 30 #How many frames does it take to go through a whole volume

#Reshape
a = img_arr[start_frame:, :, :]
new_stack = []

for i in range(a.shape[0]//frames_per_vol):
    new_stack.append(a[i*frames_per_vol:i*frames_per_vol+frames_per_vol, :, :])    

img_4d_stack = np.stack(new_stack)
```

```python
#Open volumetric data
viewer = napari.Viewer()
viewer.add_image(img_4d_stack)
```
