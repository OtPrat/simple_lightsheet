```python
from nidaqmx import Task
from nidaqmx.constants import Edge, AcquisitionType
import numpy as np
```

```python
# Rate defines the number of samples we send to the NI board (then to the Galvo) per second
task = Task()
task.ao_channels.add_ao_voltage_chan("Dev1/ao0", min_val=-10, max_val=10)
task.timing.cfg_samp_clk_timing(
    rate=40000,
    source="OnboardClock",
    active_edge=Edge.RISING,
    sample_mode=AcquisitionType.CONTINUOUS,
    samps_per_chan=1000
)
```

```python
center = 0
amp = 1

# We generate a single phase of a triangular wave with the length of 400 samples
# The signal is repeated (rate)/(length of the signal) times per second
sheet = np.concatenate(
    [
        np.linspace(-amp+center, amp+center, 200),
        np.linspace(amp+center, -amp+center, 200),
    ]
)
```

```python
task.write(sheet, auto_start=True)
```

```python
task.stop()
```
